```python
import numpy as np


def my_checker_board(n):
    M = np.ones((n, n))
    M[::2, 1::2] = 0
    M[1::2, ::2] = 0
    return M
    
print(my_checker_board(6))
```

```python

import numpy as np


def my_cylinder(r, h):
    def s(x, y):
        return 2 * np.pi * (x**2 + x * y)

    def v(x, y):
        return np.pi * y * x**2

    return [s(r, h), v(r, h)]


print(my_cylinder(2, 4))
```

```python
import numpy as np


def my_n_odds(a):
    i = 0
    suma = 0
    while i < len(a):
        if a[i] % 2 != 0:
            suma += 1
        i += 1
    return suma


print(my_n_odds(np.arange(100)))
```

```python

# VERSION 2 RECURSION

import numpy as np


def my_n_odds(a):
    if len(a) == 0:
        return 0
    else:
        return a[0] % 2 + my_n_odds(a[1:])


print(my_n_odds(np.arange(100)))
```

```python
# VERSION 3  lambda funtions
import numpy as np


def my_n_odds(a):
    return len(list(filter(lambda x: x % 2 != 0, a)))


print(my_n_odds(np.arange(100)))
```
