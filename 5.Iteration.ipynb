```python
x = [1, 2, 3, 10, 4, 5, 6, 8, 9]


def my_max(x):
    max_pos = 0
    pos = 0
    while pos < len(x):
        if x[pos] > x[max_pos]:
            max_pos = pos
        pos += 1
    return x[max_pos]


print(my_max(x))
```

```python

import numpy as np


def my_mat_mult(P, Q):

    M = np.zeros((P.shape[0], Q.shape[1]))

    for i in range(P.shape[0]):
        for j in range(Q.shape[1]):
            for k in range(Q.shape[0]):
                M[i, j] += P[i, k] * Q[k, j]

    print(M == np.dot(P, Q))
    return M


#P= np.array([[1,2],[3,4]])
#Q= np.array([[6,3],[9,5]])

P = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
Q = np.array([[1, 1, 1], [2, 2, 2], [3, 3, 3], [4, 4, 4]])
print(my_mat_mult(P, Q))
```

```python
import numpy as np


def my_trig_odd_even(M):
    Q = np.zeros((M.shape))
    for i in range(M.shape[0]):
        for j in range(M.shape[1]):
            if M[i, j] % 2 != 0:
                Q[i, j] = np.sin(np.pi / M[i, j])
            else:
                Q[i, j] = np.cos(np.pi / M[i, j])

    return Q


M = np.array([[3, 4], [6, 7]])
print(my_trig_odd_even(M))
```

```python
def my_connectivity_mat_2_dict(C, names):
    node = {}
    for i in range(len(names)):
        node[names[i]] = [[
            j + 1 for j in range(len(conection)) if conection[j] == 1
        ] for conection in C][i]
    return node


C = [[0, 1, 0, 1], [1, 0, 0, 1], [0, 0, 0, 1], [1, 1, 1, 0]]
names = ["Los Angeles", "New York", "Miami", "Dallas"]

print(my_connectivity_mat_2_dict(C, names))
```
