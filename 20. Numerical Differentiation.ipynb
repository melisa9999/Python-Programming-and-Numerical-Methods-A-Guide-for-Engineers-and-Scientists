```python
import numpy as np
import matplotlib.pyplot as plt

def f(s): return s


def my_der_calc(f, a, b, N, option):

    x = np.linspace(a, b, N)
    y = f(x)
    h = 1 / N

    if option == 'foward':
        forward_diff = np.diff(y) / h
        return forward_diff
    elif option == 'backward':
        backward_diff = (f(x + h) - f(x)) / h
        return backward_diff
    else:
        return x


print(my_der_calc(f, 1, 10, 10, 'foward'))

print(my_der_calc(f, 1, 10, 10, 'backward'))

x = np.linspace(0, np.pi, 1000)
def f(x): return np.sin(np.exp(x))


dy10 = my_der_calc(f, 0, np.pi, 1000, "backward")

plt.figure(figsize=(12, 8))
plt.plot(x, np.cos(np.exp(x)), label="analytic")
plt.plot(x, dy10, label="10 points")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
