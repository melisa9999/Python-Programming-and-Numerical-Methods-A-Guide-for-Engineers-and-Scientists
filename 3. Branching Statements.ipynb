the number of people in the group. The tip should be calculated as 15% for a party strictly less than
six people, 18% for a party strictly less than eight, 20% for a party less than 11, and 25% for a party

````python
def my_tip_calc(bill, party):
    if party < 6:
        tips = bill * 0.15
    elif party < 8:
        tips = bill * 0.18
    elif party < 11:
        tips = bill * 0.20
    elif party > 11:
        tips = bill * 0.25
    else:
        tips = "Invalid input"
    return tips


# t = 16.3935
t = my_tip_calc(109.29, 3)
print(t)

# t = 19.6722
t = my_tip_calc(109.29, 7)
print(t)

# t = 19.6722
t = my_tip_calc(109.29, 7)
print(t)

# t = 21.8580
t = my_tip_calc(109.29, 9)
print(t)

# t = 27.3225
t = my_tip_calc(109.29, 12)
print(t)

````

that is either "plus", "minus", "mult", "div", or "pow", and the function should compute: a + b,
a −b, a ∗ b, a/b, and ab for the respective values for operation. A couple of test cases are given
below.



```

In [ ]: # Output: [2,6,12,20]
```

````python
import numpy as np


def my_mult_operation(a, b, operation):
    if operation == "plus":
        out = a + b
    elif operation == "minus":
        out = a - b
    elif operation == "mult":
        out = a * b
    elif operation == "div":
        out = a / b
    elif operation == "pow":
        out = a**b
    else:
        out = "Invalid input"
    return out


x = np.array([1, 2, 3, 4])
y = np.array([2, 3, 4, 5])

# Output: [3,5,7,9]
print(my_mult_operation(x, y, "plus"))

# Output: [-1,-1,-1,-1]
print(my_mult_operation(x, y, "minus"))

# Output: [2,6,12,20]
print(my_mult_operation(x, y, "mult"))

# Output: [0.5,0.66666667,0.75,0.8]
print(my_mult_operation(x, y, "div"))

# Output: [1,8,81,1024]
print(my_mult_operation(x, y, "pow"))


````

3. Consider a triangle with vertices at (0, 0), (1, 0), and (0, 1). Write a function with the name
my_inside_triangle(x,y) where the output is the string "outside" if the point (x, y) is outside
of the triangle, "border" if the point is exactly on the border of the triangle, and "inside" if the
point is on the inside of the triangle.



```
In [ ]: def my_inside_triangle(x,y):

  return position
In [ ]: # Output: "border"
      my_inside_triangle(.5,.5)
In [ ]: # Output: "inside"
      my_inside_triangle(.25,.25)
In [ ]: # Output: "outside"
      my_inside_triangle(5,5)
```



````python

x = -5  # @param {type:"raw"}
y = 2  # @param {type:"raw"}


def my_inside_triangle(x, y):
    if (x <= 1 and x >= 0) and (y <= 1 and y >= 0):
        if x == 0 or y == 0 or round(y, 2) == round(-x + 1.0, 2):
            position = "Border"
        elif y < -x + 1:
            position = "Inside"
        else:
            position = "Outside"
    else:
        position = "Outside"
    return position


print(my_inside_triangle(x, y))

````

````python
def my_n_roots(a, b, c):
    d = b**2 - 4 * a * c
    if d > 0:
        rimplus = (-b + np.sqrt(b**2 - 4 * a * c)) / (2 * a)
        rimmin = (-b - np.sqrt(b**2 - 4 * a * c)) / (2 * a)
        return 2, (rimplus, rimmin)
    elif d == 0:
        r = -b / (2 * a)
        return 1, r
    else:
        dis = -(b**2 - 4 * a * c)
        rimplus = (-b + np.sqrt(dis) * 1j) / (2 * a)
        rimmin = (-b - np.sqrt(dis) * 1j) / (2 * a)
        return 2, (rimplus, rimmin)


# Output: n_roots = 2, r = [3, -3]
n_roots, r = my_n_roots(1, 0, -9)
n_roots, r

# Output: n_roots = -2,
# r = [-0.6667 + 1.1055i, -0.6667 - 1.1055i]
my_n_roots(3, 4, 5)

# Output: n_roots = 1, r = [1]
my_n_roots(2, 4, 2)

    

````
