```python
import numpy as np
import matplotlib.pyplot as plt



def my_double_exp(x, n):
    y = 0
    for i in range(n):
        y = y+x**(2*i)/np.math.factorial(i)
    return y, np.exp(x**2)


print(my_double_exp(2, 10))
```

```python
import numpy as np


def my_exp(x, n):
    y = 0
    for i in range(n):
        y = y+x**(i)/np.math.factorial(i)

    error = np.exp(x)*x**n/np.math.factorial(n)
    return y, np.exp(x), error


print(my_exp(2, 20))
```

```python
import numpy as np


def my_cosh(x, n):
    y = 0
    for i in range(n):
        y = y + x**(2 * i) / np.math.factorial(2 * i)

    error = np.cosh(x) * x**n / np.math.factorial(n)  # truncation error
    return y, np.cosh(x), error


print(my_cosh(0.2, 10))
```
