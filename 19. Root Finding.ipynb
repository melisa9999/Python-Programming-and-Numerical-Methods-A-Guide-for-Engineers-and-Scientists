```python
import numpy as np


def my_nth_root(x, y, n, tol):
    def f(x):
        return x**n - y

    def d_f(x):
        return n * x**(n - 1)

    if np.abs(f(x)) < tol:

        return x
    else:
        return my_nth_root(x - f(x) / d_f(x), y, n, tol)


print(my_nth_root(2.5, 9, 2, 0.1))
```

```python
def my_bisection(f, a, b, tol):

    if np.sign(f(a)) == np.sign(f(b)):
        raise Exception("The scalars a and b do not bound a root")
# get midpoint
    m = (a + b) / 2
    R.append(m)
    E.append(f(m))
    if np.abs(f(m)) < tol:
        return R, E
    elif np.sign(f(a)) == np.sign(f(m)):
        return my_bisection(f, m, b, tol)
    elif np.sign(f(b)) == np.sign(f(m)):
        return my_bisection(f, a, m, tol)


# def my_fixed_point(f,g,tol,max_iter):

# f = lambda x: x**2 - 2
# [R, E] = my_bisection(f, 0, 2, 1e-1)
# print(R)
# print(E)

def f(x): return np.sin(x) - np.cos(x)


[R, E] = my_bisection(f, 0, 2, 1e-2)

print(R)
print(E)
```

```python
import numpy as np

R = []
E = []


def my_pipe_builder(C_ocean, C_land, L, H, a, b):

    # f es la derivada del costo total C_T(x) con respecto a x
    # debe encontrarse el x que haga f=0
    def f(x):
        return x * C_ocean / (H**2 + x**2)**(1 / 2) - C_land

    if np.sign(f(a)) == np.sign(f(b)):
        raise Exception("The scalars a and b do not bound a root")
    # get midpoint
    m = (a + b) / 2
    R.append(m)
    E.append(f(m))
    if np.abs(f(m)) < 1e-6:
        return R, E
    elif np.sign(f(a)) == np.sign(f(m)):
        return my_pipe_builder(C_ocean, C_land, L, H, m, b)
    elif np.sign(f(b)) == np.sign(f(m)):
        return my_pipe_builder(C_ocean, C_land, L, H, a, m)


[R, E] = my_pipe_builder(30, 10, 100, 20, 1, 100)
print(R[-1])
print(E[-1])
```
