19. Create array x and y, where x = [1, 4, 3, 2, 9, 4] and y=[2, 3, 4, 1, 2, 3]. Compute the
assignments from Problem 3

````python
import numpy as np

x = np.array([1, 4, 3, 2, 9, 4])
y = np.array([2, 3, 4, 1, 2, 3])

u = x + y
v = x * y
w = x / y
z = np.sin(x)
r = 8 * np.sin(x)
s = 5 * np.sin(x*y)
p = x ** y

print("x =", x, "\ny =", y, "\nu =", u, "\nv =", v, "\nw =", w, "\nz =", z, "\nr =", r, "\ns =", s, "\np =", p)

````
