```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 1000)
y = 3 * np.sin(x) - 2 * np.cos(x) + np.random.random(len(x))


def func(x, a, b, c):
    f = a * np.sin(x) + b * np.cos(x) + c
    return f


def my_lin_regression(x, y):

    alpha, betha, c = optimize.curve_fit(func, xdata=x, ydata=y)[0]

    return alpha, betha, c


alpha, betha, c = my_lin_regression(x, y)
print(my_lin_regression(x, y))

plt.figure(figsize=(10, 8))
plt.plot(x, y, "b.", label="data")
plt.plot(x, alpha * np.sin(x) + betha * np.cos(x) + c, "r", label="regression")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Least Square Regression on Exponential Model")
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 1000)
y = 2 * np.exp(-0.5 * x) + 0.25 * np.random.random(len(x))


def func(x, a, b):
    y = a * np.exp(b * x)
    return y


def my_exp_regression(x, y):

    alpha, beta = optimize.curve_fit(func, xdata=x, ydata=y)[0]

    return alpha, beta


alpha, beta = my_exp_regression(x, y)
print(my_exp_regression(x, y))

plt.figure(figsize=(10, 8))
plt.plot(x, y, "b.", label="data")
plt.plot(x, alpha * np.exp(beta * x), "r", label="regression")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Least Square Regression on Exponential Model")
plt.legend()
plt.show()
```
