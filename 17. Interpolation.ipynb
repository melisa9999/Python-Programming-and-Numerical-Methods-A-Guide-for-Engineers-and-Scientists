```python
x = [1, 2, 3]
y = [4, 5, 6]
X = [2.5, 1.5]

y_hat = []


def my_lin_interp(x, y, X):

    for i in range(len(x) - 1):
        Y = y[i] + (y[i + 1] - y[i]) * (X[i] - x[i]) / (x[i + 1] - x[i])
        y_hat.append(Y)
    return y_hat


print(my_lin_interp(x, y, X))

plt.figure(figsize=(10, 8))
plt.plot(x, y, "-ob")
plt.plot(X, y_hat, "ro")
plt.title("Linear Interpolation at X")
plt.xlabel("x")

# para comprobar

f = interp1d(x, y)
y_hat = f(X)

print(y_hat)
```

```python
from scipy.interpolate import lagrange
import matplotlib.pyplot as plt
import numpy as np

x = [0, 1, 2, 3, 4]
y = [2, 1, 3, 5, 1]

X = np.linspace(0, 4, 101)


def my_lagrange(x, y, X):

    yp = 0

    for i in range(len(x)):
        p = 1

        for j in range(len(x)):
            if i != j:
                p = p * (X - x[j]) / (x[i] - x[j])

        yp = yp + p * y[i]
    return yp


# print(my_lagrange(x,y,X))

plt.figure(figsize=(10, 8))
plt.plot(X, my_lagrange(x, y, X), "b", label="interpolation")
plt.plot(x, y, "ro", label="data points")
plt.xlabel("x")
plt.ylabel("y")
plt.title(f"Lagrange Interpolation of Data Points")
plt.legend()
plt.show()


f = lagrange(x, y)
```
