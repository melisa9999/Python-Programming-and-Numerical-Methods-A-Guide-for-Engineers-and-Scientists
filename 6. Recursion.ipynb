```python
def my_sum(lst):

    if lst == []:
        return 0
    else:
        return lst.pop() + my_sum(lst)


print(my_sum([1, 2, 3]))
print(my_sum([i for i in range(1, 101)]))
```

```python
import numpy as np


def my_chebyshev_poly1(n, x):
    if n == 0:
        return [1 for i in range(len(x))]
    if n == 1:
        return x
    else:
        return 2 * np.array(x) * np.array(my_chebyshev_poly1(
            n - 1, x)) - np.array(my_chebyshev_poly1(n - 2, x))


x = [1, 2, 3, 4, 5]

print(my_chebyshev_poly1(3, x))
```

```python
def my_ackermann(m, n):
    if m == 0:
        return n + 1
    elif m > 0 and n == 0:
        return my_ackermann(m - 1, 1)
    elif m > 0 and n > 0:
        return my_ackermann(m - 1, my_ackermann(m, n - 1))


print(my_ackermann(2, 3))
```

```python
def my_n_choose_k(n, k):
    if n == k:
        return 1
    if k == 1:
        return n
    else:
        return my_n_choose_k(n - 1, k) + my_n_choose_k(n - 1, k - 1)


print(my_n_choose_k(10, 3))
```

```python
currency_denomination = [100, 50, 20, 10, 5, 1, 0.25, 0.10, 0.05, 0.01]


def my_change(cost, paid):
    change = paid - cost
    if change == 0:
        return []
    else:
        bigger_coin = [i for i in currency_denomination if i <= change][0]
        n_bigger_coin = change // bigger_coin
        paid_with_bigger_coin = bigger_coin * n_bigger_coin
        return [bigger_coin] * int(n_bigger_coin) + my_change(
            cost, round(paid - paid_with_bigger_coin, 2))


print(my_change(27.57, 100))
```

```python
def my_gcd(a, b):
    if b == 0:
        return a
    else:
        return my_gcd(b, a % b)


print(my_gcd(33, 121))
```

```python
import numpy as np


def my_spiral_ones(n, s):
    """This function calculates a nxn matrix with 0's and 1's
    The '1' numbers form a spiral.

    Examples: 
        my_spiral_ones(4,4): [[1. 1. 1. 1.]
                              [1. 0. 0. 0.]
                              [1. 0. 1. 0.]
                              [1. 1. 1. 0.]]           

    Args:
        n (int): size of the matrix
        s (int): size of the matrix (this remains constant)
    Return:
        n x n array 
    """

    if n <= 0:
        return 0
    if n == 1:
        return 1
    else:
        matrix = np.zeros((n, n))
        # left, down, right, up
        matrix[0], matrix[:, 0], matrix[-1, :-1], matrix[:1:-1, -2] = 1, 1, 1, 1
        pad_factor = s - n
        padded_matrix = np.pad(matrix, ((pad_factor // 2, pad_factor // 2),
                                        (pad_factor // 2, pad_factor // 2)),
                               constant_values=0)

        return padded_matrix + my_spiral_ones(n - 4, s)


print(my_spiral_ones(10, 10))
```
