```python
import numpy as np
import matplotlib.pyplot as plt


class Point():
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def plot_point(self):
        plt.figure(figsize=(10, 6))
        plt.plot(self.x, self.y, "-o")
        return plt.show()

    def calculate_dist(self, new_x, new_y):
        return np.sqrt((self.x - new_x)**2 + (self.y - new_y)**2)
        
first_point = Point(5, 10)
first_point.calculate_dist(20, 3)
```
